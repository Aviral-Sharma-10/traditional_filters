```python
#ANISOTROPIC
import numpy as np
from medpy.filter.smoothing import anisotropic_diffusion
import matplotlib.pyplot as plt
import cv2
import os
import skimage.io as io
import time

start_time = time.clock()
img_path="your image path"
jpg_folder_path="path where filtered image will be saved"



img=io.imread(img_path)


img_filtered1 = anisotropic_diffusion(img,1,20,0.5)
img_filtered1 = img_filtered1/float(255)

plt.imshow(img_filtered1,cmap="gray")
plt.imsave(os.path.join(jpg_folder_path, "anisotropic.png"), img_filtered1)

print((time.clock()) - start_time,"seconds")
```

```python
#BILATERAL
start_time = time.clock()
x=cv2.imread(img_path)
img_filtered2= cv2.bilateralFilter(x, 9, 75 , 75)
img_filtered2 = img_filtered2 / float(255)
plt.imshow(img_filtered2)

io.imsave(os.path.join(jpg_folder_path, "bilateral.png"), img_filtered2)
print((time.clock()) - start_time,"seconds")
```

```python
#GABOR FUNCTION

start_time = time.clock()

from skimage.filters import gabor

img_filtered_real, img_filtered_imag = gabor(img, frequency=1.5, theta = 3.14)

img_filtered_real = img_filtered_real/float(255)

io.imshow(img_filtered_real,cmap = 'gray')

# plt.imsave(("/home/dic/Desktop/filter_all/gabor.png"), img_filtered_real)

print(time.clock()-start_time)
```

```python
#LAPLACIAN
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

edges = cv2.Laplacian(img2,cv2.CV_8U,ksize=5)
                      #-1, ksize=5, scale=1, delta=0,borderType=cv2.BORDER_DEFAULT)
output = [img2, edges]

for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(output[i], cmap = 'gray')
plt.show()
```

```python
#Median 
from scipy import ndimage
start_time = time.clock()


img_filtered8 = ndimage.median_filter(img,3)

img_filtered8 = img_filtered8 / float(255)

io.imshow(img_filtered8)
print(time.clock() - start_time)
io.imsave(os.path.join(jpg_folder_path, "median.png"), img_filtered8)
```

```python
#SCHARR
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
edgesx = cv2.Scharr(img1,cv2.CV_8U,1,0)

edgesy = cv2.Scharr(img1,cv2.CV_8U,0,1)
edges = edgesx + edgesy
output = [img1, edgesx, edgesy, edges]

for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(output[i], cmap = 'gray')
    
plt.show()
```

```python
#SOBEL
edgesx =  cv2.Sobel(img,cv2.CV_8U,1,0,ksize=5)

edgesy = cv2.Sobel(img,cv2.CV_8U,0,1,ksize=5)

edges = edgesx + edgesy

output = [img, edgesx, edgesy, edges]
    
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(output[i], cmap = 'gray')
plt.show() 
```

```python
#WAVELET
import skimage
start_time=time.clock()
img_filtered9 = skimage.restoration.denoise_wavelet(img, 0.5, wavelet='sym9')
# img_filtered9 = img_filtered9 / float(255)

io.imshow(img_filtered9)

io.show()
print(time.clock() - start_time)
io.imsave(os.path.join(jpg_folder_path, "wavelet.png"), img_filtered9)
```
